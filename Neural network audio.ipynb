```python
import json
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras

import matplotlib.pyplot as plt

# path to json file that stores MFCCs and genre labels for each processed segment
data_path = "/Users/LENOVO/Desktop/Programming/Neural network/Neural network/mfccatis.json"
```

```python
with open(data_path, "r") as fp:
    data = json.load(fp)
    
# convert lists to numpy arrays
allLabels = np.array(data["labels"])
allData = np.array(data["mfcc"])

allData.shape
```

```python
# create train/test split, using sklearn
dataTrain, dataTest, labelTrain, labelTest = train_test_split(allData, allLabels, test_size=0.3)
```

```python
dataTrain.shape 
```

```python
# build network topology
model = keras.Sequential([

    # input layer
    keras.layers.Flatten(input_shape=(dataTrain.shape[1], dataTrain.shape[2])),
    
    # 1st dense layer
    #keras.layers.Dense(195 , activation='relu', kernel_regularizer=keras.regularizers.l2(0.001)),
    #keras.layers.Dropout(0.1),

    # 2nd dense layer
    #keras.layers.Dense(145, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001)),
    #keras.layers.Dropout(0.1),

    #3rd dense layer
    #keras.layers.Dense(64, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001)),
    #keras.layers.Dropout(0.1),

    # output layer
    keras.layers.Dense(10, activation='relu')
])
# compile model
optimiser = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=optimiser, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(dataTrain, labelTrain, validation_data=(dataTest, labelTest), batch_size=32, epochs=300)
```

```python
fig, axs = plt.subplots(2)

# create accuracy sublpot
axs[0].plot(history.history["accuracy"], label="train accuracy")
axs[0].plot(history.history["val_accuracy"], label="test accuracy")
axs[0].set_ylabel("Accuracy")
axs[0].legend(loc="lower right")
axs[0].set_title("Accuracy eval")

# create error sublpot
axs[1].plot(history.history["loss"], label="train error")
axs[1].plot(history.history["val_loss"], label="test error")
axs[1].set_ylabel("Error")
axs[1].set_xlabel("Epoch")
axs[1].legend(loc="upper right")
axs[1].set_title("Error eval")

plt.show()
```
